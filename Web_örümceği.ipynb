```python

import requests #hazır olan kütüphaneye kodu dahil edip istek gönderir
url = input("url giriniz")
response = requests.get(url) # kütüphanesini kullanarak URL'ye bir HTTP GET isteği gönderir.kaynaktan veri almak için kullanılır
print(response.text) #yanıtın içeriğini html olarak alır
```

```python
import requests
from bs4 import BeautifulSoup

# Kullanıcıdan URL alma
url = input("URL giriniz: ")

# GET isteği gönderme
response = requests.get(url)

# Durum kodunu kontrol etme
if response.status_code == 200:
    # BeautifulSoup ile sayfanın HTML içeriğini çözümleme
    soup = BeautifulSoup(response.content, 'html.parser')

    # Tüm bağlantıları bul
    links = soup.find_all('a')  # 'a' etiketleri ile bağlantıları bul

    # Bağlantı metinlerini yazdır
    for link in links:
    print(link.get_text())
else:
    print('Sayfa alınamadı:', response.status_code)  # Hata durum kodunu yazdır
```
